```python
### Reflecting DB
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import func
from sqlalchemy import and_, or_, not_
import pandas as pd
import numpy as np
```

```python
# Create the engine to connect to a sqlite file.
engine = create_engine("sqlite:///CCES_Ver61.sqlite")

# Declare a base using automap_base()
Base = automap_base()


Base.prepare(engine, reflect=True)

# Print all of the classses mapped to the Base (this provides a list of the table names in the sqlite.db)
Base.classes.keys()
```

```python
# Assign the CCES_16 class to a variable called 'CCES'
CCES = Base.classes.Cces_16
```

```python
# Create a variable to manage session.queries later on...
session = Session(engine)
# session = Session(bind = engine)
```

```python
#############################################
# Dropdown Variables mapped to Sql...
#############################################
 
# Gun Questions
GunBackgroundChecks_16 = CCES.GunBackgroundChecks_16
ProhibitPublication_16 = CCES.ProhibitPublication_16
BanAssultWeapons_16 = CCES.BanAssultWeapons_16
MakeCCPEasier_16 = CCES.MakeCCPEasier_16

# Abortion Questions
AlwaysAllowChoice_16 = CCES.AlwaysAllowChoice_16
RapeIncestorHealth_16 = CCES.RapeIncestorHealth_16
ProhibitMoreThan20Weeks_16 = CCES.ProhibitMoreThan20Weeks_16
Employersdeclinebenefits_16 = CCES.Employersdeclinebenefits_16
ProhibitFedFunds_16 = CCES.ProhibitFedFunds_16

# Gay Marriage Question
GayMarriage = CCES.GayMarriage
#############################################
```

```python
# GunBackgroundChecks
#Query
sample = GunBackgroundChecks_16
StatesFor = session.query(CCES.StateAbb, func.count(sample)).\
                filter(sample == 'Support').\
                group_by(CCES.StateAbb).all()

StatesTotal = session.query(CCES.StateAbb, func.count(sample)).\
                filter(sample != 'nothin').\
                group_by(CCES.StateAbb).all()

#DF Generation
GunBackgroundChecks1 = pd.DataFrame(StatesFor, columns=['StateAbb', 'VotesFor'])
GunBackgroundChecks2 = pd.DataFrame(StatesTotal, columns=['StateAbb', 'TotalVotes'])
display(GunBackgroundChecks1.head())
display(GunBackgroundChecks2.head())
GunBackgroundChecksjoined = pd.merge(GunBackgroundChecks1, GunBackgroundChecks2, on='StateAbb')
GunBackgroundChecks_16 = GunBackgroundChecksjoined.VotesFor/ GunBackgroundChecksjoined.TotalVotes
GunBackgroundChecksjoined['GunBackgroundChecks_16']=GunBackgroundChecks_16
#Final DF
GunBackgroundChecksFinal = GunBackgroundChecksjoined[['StateAbb', 'GunBackgroundChecks_16']]
GunBackgroundChecksFinal.head()
```
